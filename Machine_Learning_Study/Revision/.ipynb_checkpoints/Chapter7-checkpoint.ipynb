# Chapter 7: Ensemble Learning and Random Forest

1. Ensemble Learining:
    - aggregate the predictions of a group of predictors (such as classifiers or regressors), you will often get better predictions than with the best individual predictor
    - A group of predictors is called an ensemble; 
    
2.  Random Forest
    - train a group of Decision Tree classifiers, __each on a different random subset__ of the training set
    - To make predictions, you ___obtain the predictions of all the individual trees__
    - then predict the class that gets the most votes
    -  will often use Ensemble methods near the end of a project, once you have already built a few good predictors

3. Popular Ensemble Learning methods:
    - bagging, boosting, and stacking. 
    - Random Forests

# Voting Class

Suppose you already have a few classifier, each about 80% accuracy
- __Hard voting__ classifier: predict the class that get most vote. __mojority-vote classifier__ (better acurracy than individual)
- If each indivudual classifier is a __weak learner__(only slightly better than random gussing), then ensemble can be __strong learner__
  - Only if there are __sufficient number of weak learners__ and they are sufficiently __diverse__
  - making the uncorrelated errors

```python
heads_proba = 0.51
coin_tosses = (np.random.rand(10000, 10) < heads_proba).astype(np.int32)
# np.cumsum: Return the cumulative sum of the elements along a given axis.
cumulative_heads_ratio = np.cumsum(coin_tosses, axis=0) / np.arange(1, 10001).reshape(-1, 1)
```

```python
plt.figure(figsize=(8,3.5))
plt.plot(cumulative_heads_ratio)
plt.plot([0, 10000], [0.51, 0.51], "k--", linewidth=2, label="51%")
plt.plot([0, 10000], [0.5, 0.5], "k-", label="50%")
plt.xlabel("Number of coin tosses")
plt.ylabel("Heads ratio")
plt.legend(loc="lower right")
plt.axis([0, 10000, 0.42, 0.58])

plt.show()
```

## Make independent predictors
- The ensemble method works best when predictors are independent 
- one way to get __diverse classifier__:
  - train them using very different algorithms
  - make different types of errors
  
## Using Scikit learning to trains a voting classifiers

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

# using the moons datasets with noise
X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Training the dataset using
log_clf = LogisticRegression(solver="lbfgs", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)
svm_clf = SVC(gamma="scale", random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='hard')
```

```python
voting_clf.fit(X_train, y_train)
```

## Accuracy score on test set

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

__The voting classifier has better performance__

## Soft voting
- instead of hard voting (output the most votted class by predictors_
- this, each predictors calculate __probability__ for each class
- output the highest class probabilities
- better than hard voting method
- for SVC: set ___probability___ hyperparameter true. 

```python
log_clf = LogisticRegression(solver="lbfgs", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)
svm_clf = SVC(gamma="scale", probability=True, random_state=42)

## set voting "soft"
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft')
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

# Bagging and Pasting
__To get diverse classifier__:

- Use very different training algorithm
- __Same training algorithm__ for every predictor and train them on __different random subsets__ of training set
  
  - __Bagging__: Sampling performed ___with___ replacement
  - __Pasting__: Sampling performed ___without___ replacement
  
- allow training instances to be sampled several times across multiple predictors

- only bagging allows training instances to be sampled several times for the same predictor.

## Bagging and Pasting in Scikit-Learn
- BaggingClassifier or BaggingRegressor
- If want to use pasting, set __bootstrap=False__
- __n_jobs__ parameter tells Scikit-learn the number of __CPU cores__ to use for training and predicitions
- __n_jobs=-1__ : to use all available cores

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

## Ensembles 500 DecisionTreeClassifier
# Each is trained on 100 training instances ramdomly sampled with replacement
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1, random_state=42)
bag_clf.fit(X_train, y_train)
y_pred = bag_clf.predict(X_test)

# pasting: set bootstrap False
pasting_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators=500,
    max_samples=100, bootstrap=False, n_jobs=-1, random_state=42)
pasting_clf.fit(X_train, y_train)
y_pred_past = pasting_clf.predict(X_test)
```

```python
# Accuracy score for bagging method
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```

```python
# accuracy of a single Decision Tree
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
y_pred_tree = tree_clf.predict(X_test)
print(accuracy_score(y_test, y_pred_tree))
```

__Example of Comparison__
- Single Decision Tree vs. 500 Trees ensembles forests
- ensemble's generalize much better . It has comparable bias but smaller variance
- The same number of errors on training set, but __decision  boundary less irregular__

```python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(clf, X, y, axes=[-1.5, 2.45, -1, 1.5], alpha=0.5, contour=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s) # Return coordinate matrices from coordinate vectors.
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)
    if contour:
        custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
        plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", alpha=alpha)
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", alpha=alpha)
    plt.axis(axes)
    plt.xlabel(r"$x_1$", fontsize=18)
    plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
```

```python
fix, axes = plt.subplots(ncols=3, figsize=(15,4), sharey=True)
plt.sca(axes[0])
plot_decision_boundary(tree_clf, X, y)
plt.title("Decision Tree", fontsize=14)
#Set the current Axes instance to axes 1
#The current Figure is updated to the parent of axes1.
plt.sca(axes[1])
plot_decision_boundary(bag_clf, X, y)
plt.title("Decision Trees with Bagging", fontsize=14)
plt.ylabel("")

plt.sca(axes[2])
plot_decision_boundary(pasting_clf, X, y)
plt.title("Decision Trees with Pasting", fontsize=14)
plt.ylabel("")

plt.show()
```

__Bootstrapping introduces more diversity__
- Bagging ends up wth __higher bias__ than pasting
- but extra diversity means predictors are less correlated, so ensemble's __variance__ is reduced

# Out-of-Bag Evaluation
- some instances may be sampled several times for any given predictor, while others may not be sampled at all. 
- only about 63% of the training instances are sampled on average for each predictor
- The remaining 37% of the training instances that are not sampled are called __outof-bag (oob) instances__.
- We can do validation on this 37% data without separate of validation set
- Evaluate the ensembles by __averaging out the oob evaluations of each predictors__

```python
# Set oob_score to True to enable automatic oob evaluation after training
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators=500,
    bootstrap=True, oob_score=True, random_state=40)
bag_clf.fit(X_train, y_train)
bag_clf.oob_score_
```

 so according to this oob evaluation, the classifier is likely to have 90% accuracy on test set
 
 - Verification:

```python
from sklearn.metrics import accuracy_score
y_pred = bag_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

# Random Patches and Random Subspace

1. BaggingClassifier support __sampling the features__
   - two hyperparameters: __max_features__ and __bootstrap_features__
   - each predictors will be trained on __random subset__ of the input features
   - Particularly useful for __high-dimensional inputs__ such as images

2. Sampling both training instances and features is called __Random Patches method__

3. __Random Subspace method__:
   - Keeping all training instances (by setting bootstrap=False and max_samples=1.0) but __sampling only features__ (by setting bootstrap_features to True and/or max_features to a value smaller than 1.0) 
   
4. Sampling features results in __even more predictor diversity__, trading a bit more bias for a lower variance.

# Boosting (Hypothesis boosting)

- Combine several weak learners into strong learner
- Triain the predictors __sequentially__, try to correct its predecessor
- __AdaBoost__(adaptive boosting) and __Gradient Boosting__

## AdaBoost
- Let the __next predictor__ focus on Training instances that the predecessor ___underfitted___
- Add __Sample Wighet__ for misclassified instance
- After training, ensemble makes predictions with __different weight__ of predictions, which based on their __overall accuracy__ on the weighted training set.

- E.g
  - first train a __base classifier__, make its predictions on training set
  - Increase relative weight of __misclassified__ training instances
  - Trains second classifier using updated weights
  
### Decision Boundaries of consecutive predictors
--> __making better predictions each time__

```python
# Each predictor is a highly regularized SVM classifier with an RBF kernel
m = len(X_train)

fix, axes = plt.subplots(ncols=2, figsize=(10,4), sharey=True)
for subplot, learning_rate in ((0, 1), (1, 0.5)):
    sample_weights = np.ones(m)
    plt.sca(axes[subplot])
    for i in range(5):
        svm_clf = SVC(kernel="rbf", C=0.05, gamma="scale", random_state=42)
        svm_clf.fit(X_train, y_train, sample_weight=sample_weights)
        y_pred = svm_clf.predict(X_train)
        
        # Sample weight get boosted for wrong predictions
        sample_weights[y_pred != y_train] *= (1 + learning_rate)
        plot_decision_boundary(svm_clf, X, y, alpha=0.2)
        plt.title("learning_rate = {}".format(learning_rate), fontsize=16)
    if subplot == 0:
        plt.text(-0.7, -0.65, "1", fontsize=14)
        plt.text(-0.6, -0.10, "2", fontsize=14)
        plt.text(-0.5,  0.10, "3", fontsize=14)
        plt.text(-0.4,  0.55, "4", fontsize=14)
        plt.text(-0.3,  0.90, "5", fontsize=14)
    else:
        plt.ylabel("")

plt.show()
```

# Gradient Boosting
- Also sequentially adding predictors to ensemble
- This method tries to fit new predictor to ___residual errors___ made by previous one

###  Gradient Tree Boosting or Gradient Boosted Regression Tree __(GBRT)__ 
- Gradient boosting using Decision Tree as base class

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Train the first DecisionTree
tree_reg1 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg1.fit(X, y)
```

```python
# Residul error made by the first predictor
y2 = y - tree_reg1.predict(X) 

# Second Decision Tree Trained Based on residual error
tree_reg2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg2.fit(X, y2)
```

```python
# Train the third predictor
y3 = y2 - tree_reg2.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg3.fit(X, y3)
```

```python
# Adding up the predictions of all the trees
# Make predictions on new instance
X_new = np.array([[0.8]])
y_pred = sum(tree.predict(X_new) for tree in (tree_reg1, tree_reg2, tree_reg3))
```

```python
y_pred
```

### Plot the above result

```python
def plot_predictions(regressors, X, y, axes, label=None, style="r-", data_style="b.", data_label=None):
    x1 = np.linspace(axes[0], axes[1], 500)
    y_pred = sum(regressor.predict(x1.reshape(-1, 1)) for regressor in regressors)
    plt.plot(X[:, 0], y, data_style, label=data_label)
    plt.plot(x1, y_pred, style, linewidth=2, label=label)
    if label or data_label:
        plt.legend(loc="upper center", fontsize=16)
    plt.axis(axes)
```

```python
plt.figure(figsize=(11,11))

# predictions of these three trees in the left column
#  ensemble’s predictions in the right column
plt.subplot(321)
plot_predictions([tree_reg1], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h_1(x_1)$", style="g-", data_label="Training set")
plt.ylabel("$y$", fontsize=16, rotation=0)
plt.title("Residuals and tree predictions", fontsize=16)

plt.subplot(322)
plot_predictions([tree_reg1], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h(x_1) = h_1(x_1)$", data_label="Training set")
plt.ylabel("$y$", fontsize=16, rotation=0)
plt.title("Ensemble predictions", fontsize=16)

plt.subplot(323)
plot_predictions([tree_reg2], X, y2, axes=[-0.5, 0.5, -0.5, 0.5], label="$h_2(x_1)$", style="g-", data_style="k+", data_label="Residuals")
plt.ylabel("$y - h_1(x_1)$", fontsize=16)

plt.subplot(324)
plot_predictions([tree_reg1, tree_reg2], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h(x_1) = h_1(x_1) + h_2(x_1)$")
plt.ylabel("$y$", fontsize=16, rotation=0)

plt.subplot(325)
plot_predictions([tree_reg3], X, y3, axes=[-0.5, 0.5, -0.5, 0.5], label="$h_3(x_1)$", style="g-", data_style="k+")
plt.ylabel("$y - h_1(x_1) - h_2(x_1)$", fontsize=16)
plt.xlabel("$x_1$", fontsize=16)

plt.subplot(326)
plot_predictions([tree_reg1, tree_reg2, tree_reg3], X, y, axes=[-0.5, 0.5, -0.1, 0.8], label="$h(x_1) = h_1(x_1) + h_2(x_1) + h_3(x_1)$")
plt.xlabel("$x_1$", fontsize=16)
plt.ylabel("$y$", fontsize=16, rotation=0)


plt.show()
```

## Finding optimal number of trees- Early stopping
### First approach:  __Staged_predict()__ method
- return the iterator over predictions
- measures __validation error__ at each stage, then trains another GBRT ensemble using optimal number
- training a large number of trees first and then looking back to find theoptimal number

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=49)

gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=120, random_state=42)
gbrt.fit(X_train, y_train)

# compute MSE of prediction
errors = [mean_squared_error(y_val, y_pred)
          for y_pred in gbrt.staged_predict(X_val)]

# best estimator is the one after yeild minmum error
bst_n_estimators = np.argmin(errors) + 1

gbrt_best = GradientBoostingRegressor(max_depth=2, n_estimators=bst_n_estimators, random_state=42)
gbrt_best.fit(X_train, y_train)
```

```python
min_error = np.min(errors)

plt.figure(figsize=(10, 4))

plt.subplot(121)
plt.plot(errors, "b.-")
plt.plot([bst_n_estimators, bst_n_estimators], [0, min_error], "k--")
plt.plot([0, 120], [min_error, min_error], "k--")
plt.plot(bst_n_estimators, min_error, "ko")
plt.text(bst_n_estimators, min_error*1.2, "Minimum", ha="center", fontsize=14)
plt.axis([0, 120, 0, 0.01])
plt.xlabel("Number of trees")
plt.ylabel("Error", fontsize=16)
plt.title("Validation error", fontsize=14)

plt.subplot(122)
plot_predictions([gbrt_best], X, y, axes=[-0.5, 0.5, -0.1, 0.8])
plt.title("Best model (%d trees)" % bst_n_estimators, fontsize=14)
plt.ylabel("$y$", fontsize=16, rotation=0)
plt.xlabel("$x_1$", fontsize=16)

plt.show()
```

### Second approach: incremental training
- __warm_start=True__
- stop training when validation erro __does not imporve for five iteration__ in a row

```python
gbrt = GradientBoostingRegressor(max_depth=2, warm_start=True, random_state=42)

min_val_error = float("inf")
error_going_up = 0
for n_estimators in range(1, 120):
    gbrt.n_estimators = n_estimators
    gbrt.fit(X_train, y_train)
    y_pred = gbrt.predict(X_val)
    val_error = mean_squared_error(y_val, y_pred)
    if val_error < min_val_error:
        min_val_error = val_error
        error_going_up = 0
    else:
        error_going_up += 1
        if error_going_up == 5:
            break  # early stopping
```
